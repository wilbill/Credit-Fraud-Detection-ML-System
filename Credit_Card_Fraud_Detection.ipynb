```python
#Dataset got from https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
#these are dependencies
import numpy as np # used for making arrays
import pandas as pd # data is in csv => hard to process & analyze, hence in pandas help to do data-frames w/c are easier to do analysis and processing (data frames are structured table)
from sklearn.model_selection import train_test_split #train_test_split fn helps in spliting data into traing & test data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #Helps in checking perfomance of our model
```

```python
#Loading the dataset to a pandas dataFrame
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
#Checking the distribution of legit transaction and fraudulent transcations, 0=> normal transcation, 1=> Fraud
credit_card_data['Class'] .value_counts()  #Class is the real column name---other cols are v26, amount etc
```

```python
#its clear data-set is unbalanced....look at fraudulent and normal transactions?
```

```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class==0] #Making only legit DataFrame
fraud = credit_card_data[credit_card_data.Class==1] #Making a fraud DataFrame
print(legit.shape)
print(fraud.shape)
```

```python
#Statistic measures of the data for a single column
legit.Amount.describe() #just legit stat of 1 column, we can do other columns like Time, Class, V20 etc
```

```python
#NB: The 25% is a percentile, so 25%==5.650000 means 25 percent of the transactions are less than 5.65 dollars
```

```python
#Now fraud stats for the Amount col
fraud.Amount.describe()
```

```python
#Comparing the mean values for all cols of fraud and legit transactions
credit_card_data.groupby('Class').mean()
```

```python
#Dealing with unbalanced data,
#we need to take some small sample from the legit data
# No. of fraudulent Transactions --> 492
# No. of non-fraudulent Transactions -->284315
#I will take a small sample of legit DataFrame (i.e, 492) transactions to match the 492 fraudulent==> balancing
legit_sample = legit.sample(n=492) #n = number of values we want, we took a sample of 492 data points and it is random


#Concatenating two DataFrames (remember fraud is the whole of it while legit used here is a random sample)
new_dataset = pd.concat([legit_sample, fraud], axis=0) #axis=0 means values added row-wise and added 1 by 1 and all fraud will be added below the legit_sample, axis=1 means values added column-wise.
```

```python
#Checking first 5 rows of the new dataset
new_dataset.head()
```

```python
#Checking the last 5 rows of the new data-set
new_dataset.tail()
```

```python
#Now checking the DIstribution, but on new data set
new_dataset['Class'].value_counts()
```

```python
#Comparing the mean values for both fraud and legit cols for all cols
#we selected col called class and found the mean, class has 0 and 1, so we get mean for all cols with class=0 and those =1
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into features and Targets.
#X will represent all the columns except the dropped one, ie Class Column
X = new_dataset.drop(columns='Class', axis=1) #axis=1 represents col
Y= new_dataset['Class']

print(X) #total=30 coz last column has been dropped
print("WE ARE NOW PRINTING FOR Y\n")
print(Y) #Y will be a variable to store the values of the Class column that X dropped, it contains the 0 or 1 (Labels)
```

```python
#Splitting the data into Training data and Testing data
#test_size = 0.2 implies 20% of test_data, 80% goes to training data. test data is just a small percentage
#we have 4 arrays here
#X_train=> varibale to hold data to be trained
#Stratify = Y, to make an even distribution of data, both in X_train and X_test, if not mentioned, distribution of 0 and 1 can be different in training and testing data
#random_state=2 or 3, is a way how data will be distributed, ie reproduce code the way it is
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#I have successfully split the data into training and test_data
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING STEP


```python
#model = LogisticRegression() # this failed, I used a different solver, also max_iter=500 failed.
model = LogisticRegression(solver='liblinear') #I used a different solver,
model.fit(X_train, Y_train)
```

MODEL EVALUATION AKA ACCURACY_SCORE


```python
#Accuracy on Training data, ie, testing the model with the data it already knows, like solving numbers it has already seen
X_train_prediction = model.predict(X_train) #predicting the labels
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) #

#Printing accuracy on training Data
print('Accuracy on Training Data: ', training_data_accuracy) #0.75+ is pretty good prediction, 0.94 means, out of 100, model will predict 94 times correct
```

```python
#Now testing the model with data it has never seen, this is like an exam for the model, testing on numbers it has never seen
#Accuracy on Test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Now Accuracy Score on Test Data ', testing_data_accuracy) #Wow, 92.4% is pretty good, model learnt well
```

```python
#Why did I get accuracy score on training data,
#Answer => I needed to compare accuracy score on test and train because a significant difference means either
#underfitting or overFitting can occur is what is termed when the difference is big,
#if training accuracy > testing significantly, means we over-fitted the model
#ie, if test is better than train significantly => We have under-fitted the model
```
